```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')

def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Tokenize the text into words
    words = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a single string
    processed_text = ' '.join(filtered_words)

    return processed_text

```

```python
# Apply the preprocess_text function to the "SYMPTOM_TEXT" column
covid_data['SYMPTOM_TEXT'] = covid_data['SYMPTOM_TEXT'].apply(preprocess_text)
```

```python
from tqdm import tqdm
symptoms_to_entities = {}
for idx, doc in tqdm(enumerate(covid_data["SYMPTOM_TEXT"])):
    out = nlp(doc)
    inner_dic = {}
    for ent in out.entities:
        if ent.type not in inner_dic:
            inner_dic[ent.type] = []
        inner_dic[ent.type].append(ent.text)
    symptoms_to_entities[covid_data['VAERS_ID'][idx]] = inner_dic
```

```python
from collections import Counter

# Extract all unique problems
all_problems = [problem for entity_data in symptoms_to_entities.values() for problem in entity_data.get('PROBLEM', [])]

# Calculate the count of each symptom
symptom_counts = Counter(all_problems)

# Filter symptoms with a count above the threshold (200)
threshold = 200
most_frequent_symptoms = {symptom: count for symptom, count in symptom_counts.items() if count > threshold}

# Display the results
print("All unique problems:", set(all_problems))
print("\nSymptom counts:", symptom_counts)
print("\nMost frequent symptoms (count > 500):", most_frequent_symptoms)
```

```python
# Plotting
symptoms, counts = zip(*most_frequent_symptoms.items())

plt.bar(symptoms, counts)
plt.xlabel('Symptoms')
plt.ylabel('Count')
plt.title('Most Frequent Symptoms (Count > 200)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
import csv
# Export to CSV
csv_file_path = 'all_unique_problems.csv'
with open(csv_file_path, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Problem'])
    csv_writer.writerows([[problem] for problem in all_problems])

print(f"Data exported to {csv_file_path}")
```

```python
symptom_frequency = {}

for symptom in all_problems:
    symptom = symptom.lower()
    symptom = symptom.strip("''")
    symptom_frequency[symptom] = symptom_frequency.get(symptom, 0) + 1

# Sort the dictionary by frequency in descending order
sorted_symptoms = {k: v for k, v in sorted(symptom_frequency.items(), key=lambda item: item[1], reverse=True)}

print(sorted_symptoms)
```

```python
csv_file_path = 'symptom_frequency.csv'

with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = ['Symptom', 'Frequency']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for symptom, frequency in sorted_symptoms.items():
        writer.writerow({'Symptom': symptom, 'Frequency': frequency})

print(f'The CSV file has been created at: {csv_file_path}')
```
